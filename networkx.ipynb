```python
degrees = np.array([G.in_degree[i] for i in range(200)])
M= degrees.max()
histo=np.histogram(degrees,bins=M+1,range=(-0.1,M+.1))
binc = np.arange(0,M+1)

fig2,ax2=plt.subplots()
#ax2.hist(degrees,bins=M, range=(-0.1,M+.1))
ax2.plot(binc,histo[0]/histo[0].sum(),'.')
#ax2.set_yscale('log')
ax2.grid()
```
